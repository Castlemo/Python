## ✅ 마크다운 기본 문법 요약표

| 기능             | 문법 예시                                   | 결과 예시                     |
|------------------|----------------------------------------------|-------------------------------|
| **제목**         | `# 제목1`<br>`## 제목2`<br>`### 제목3`       | 제목1 / 제목2 / 제목3         |
| **굵은 글씨**    | `**굵게**` or `__굵게__`                     | **굵게**                      |
| *기울임 글씨*     | `*기울임*` or `_기울임_`                    | *기울임*                      |
| **코드 블럭**    | `` `코드` ``                                | `코드`                        |
| **여러 줄 코드** | <pre>```<br>코드<br>```</pre>               | 코드 블럭으로 표시됨          |
| **링크**         | `[이름](https://example.com)`                | [이름](https://example.com)   |
| **이미지**       | `![대체텍스트](https://image.url)`           | 이미지 표시됨                 |
| **리스트**       | `- 항목1`<br>`- 항목2`                       | - 항목1<br>- 항목2            |
| **번호 리스트**  | `1. 항목`<br>`2. 항목`                       | 1. 항목<br>2. 항목            |
| **수평선**       | `---` or `***`                               | ―――                           |
| **인용문**       | `> 인용문`                                   | > 인용문                      |
| **체크박스**     | `- [ ] 할 일`<br>`- [x] 완료`                | - [ ] 할 일<br>- [x] 완료     |
| **표 만들기**    | `| A | B |`<br>`|---|---|`<br>`| 1 | 2 |`     | A \| B<br>1 \| 2              |
